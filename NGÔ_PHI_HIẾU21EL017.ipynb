```python
#vẽ biểu đồ tròn
#copy right by Phi Hieu
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Housing.csv')
print(df)
```

```python
# Vẽ biểu đồ tròn
# Data for Furnishing Status
furnishing_counts = df['furnishingstatus'].value_counts()
# Data for Main Road
mainroad_counts = df['mainroad'].value_counts()

# Set up the figure
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Pie chart for Furnishing Status
axs[0].pie(furnishing_counts, labels=furnishing_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'salmon'])
axs[0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
axs[0].set_title('Distribution of Furnishing Status', fontsize=14)

# Pie chart for Main Road
axs[1].pie(mainroad_counts, labels=mainroad_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightyellow'])
axs[1].axis('equal')
axs[1].set_title('Houses on Main Road', fontsize=14)

# Show the plots
plt.tight_layout()
plt.show()
```

```python
# Vẽ biểu đồ đường
# Grouping by area and calculating the average price
average_price_by_area = df.groupby('area')['price'].mean().reset_index()

# Sorting the data by area for the line chart
average_price_by_area.sort_values('area', inplace=True)

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Plotting the line chart
plt.figure(figsize=(12, 6))
sns.lineplot(data=average_price_by_area, x='area', y='price', marker='o', color='teal', linewidth=2, markersize=5)

# Adding titles and labels
plt.title('Biểu đồ đường', fontsize=16)
plt.xlabel('Area (Square Feet)', fontsize=14)
plt.ylabel('Average Price (VND)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
